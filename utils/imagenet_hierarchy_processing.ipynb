```python
import pandas as pd
import json
```

```python
def get_superclass_map(class_dict, tree_data, superclass_level):
#     classes = set(map(lambda s : s.strip().lower(), class_dict.values()))
    classes = {cls:idx for idx, cls in class_dict.items()}
#     print(classes)
    res_dict = {}
    curr_supernodes = []
    def get_superclass_recursive(curr_dict):
        if "children" in curr_dict:
            if curr_dict['name'] != "ImageNet 2011 Fall Release":
                curr_supernodes.append(curr_dict['name'])

            for node in curr_dict['children']:
                get_superclass_recursive(node)

                
            if curr_dict['name'] != "ImageNet 2011 Fall Release":
                del curr_supernodes[-1]
        
#         this_node_names = curr_dict['name'].split(",")
        
#         for node_name in this_node_names:
# #             print(node_name.strip().lower())
#             if node_name.strip().lower() in classes:
#                 res_dict[node_name.strip().lower()] = curr_supernodes[-3 if len(curr_supernodes) >= 3 else  0].split(",")[0].strip().lower()
        if curr_dict['name'] in classes:
                res_dict[classes[curr_dict['name']]] = (curr_dict['name'].split(",")[0].strip().lower() if len(curr_supernodes) == 0 else (curr_supernodes[-2 if len(curr_supernodes) >= 2 else  0].split(",")[0].strip().lower()))
                
    get_superclass_recursive(tree_data)
    return res_dict
```

```python
data = json.load(open("imagenet_hierarchy_data.json", "r"))
data
```

```python
imagenet_df = pd.read_csv("imagenet_class2.csv")
class_dict = {i: cls for i, cls in enumerate(imagenet_df['Class Name'])}
class_dict
```
